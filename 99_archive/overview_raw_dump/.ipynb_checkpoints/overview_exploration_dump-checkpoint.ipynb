```python
import numpy as np
import pandas as pd
from src.utils.project_root import get_project_root

# ---------------------------------------------------------------------
# 0. Paths
# ---------------------------------------------------------------------
ROOT = get_project_root()
DATA_RAW = ROOT / "data" / "raw"
DATA_RAW.mkdir(parents=True, exist_ok=True)

np.random.seed(13)

regions = ["austin", "toronto", "shanghai", "hong_kong"]
dates = pd.date_range("2010-01-01", "2025-12-01", freq="MS")

rows = []

for region in regions:
    n = len(dates)

    # --- region-specific starting levels ---
    if region == "austin":
        income0, pop0, price0 = 60000, 1.8e6, 250000
    elif region == "toronto":
        income0, pop0, price0 = 72000, 2.8e6, 400000
    elif region == "shanghai":
        income0, pop0, price0 = 150000, 20e6, 300000
    else:  # hong_kong
        income0, pop0, price0 = 200000, 7.5e6, 500000

    income = [income0]
    pop = [pop0]
    net_migration = [3000]

    # --- income / population paths ---
    for t in range(1, n):
        # income: 3% trend + cycle + noise
        g_income_trend = 0.03 / 12
        g_income_cycle = 0.01 * np.sin(2 * np.pi * t / 60)
        g_income = g_income_trend + g_income_cycle + np.random.normal(0, 0.005)
        income.append(income[-1] * (1 + g_income))

        # population: ~1%/yr growth + noise
        g_pop = 0.01 / 12 + np.random.normal(0, 0.001)
        pop.append(pop[-1] * (1 + g_pop))

        # net migration roughly proportional to population
        nm = 0.0015 * pop[-1] + np.random.normal(0, 1000)
        net_migration.append(max(nm, -5000))

    # --- global-ish mortgage rate regime path (same for all regions) ---
    mort = []
    for d in dates:
        year = d.year
        if year <= 2014:
            base = 0.045  # pre-QE exit
        elif year <= 2019:
            base = 0.040
        elif year <= 2021:
            base = 0.025  # covid lows
        else:
            base = 0.055  # hiking cycle
        mort.append(base + np.random.normal(0, 0.002))
    mort = np.clip(mort, 0.015, 0.08)

    income = np.array(income)
    pop = np.array(pop)
    net_migration = np.array(net_migration)
    mort = np.array(mort)

    # --- rents: region-specific rent-to-income ratio ---
    rent_mult = {
        "austin": 0.28,
        "toronto": 0.35,
        "shanghai": 0.32,
        "hong_kong": 0.45,
    }[region]
    rent = rent_mult * income / 12 * (1 + np.random.normal(0, 0.05, size=n))

    # --- permits: proxy for new supply ---
    permits = 0.001 * pop + np.random.normal(0, 100, size=n)
    permits = np.clip(permits, 0, None)

    # --- inventory: months of supply ---
    inventory = np.random.normal(6, 1, size=n)
    inventory = np.clip(inventory, 2, 15)

    # --- unemployment: simple business cycle ---
    unemployment = (
        0.06
        + 0.02 * np.sin(2 * np.pi * np.arange(n) / 84)
        + np.random.normal(0, 0.005, size=n)
    )
    unemployment = np.clip(unemployment, 0.02, 0.15)

    # --- house prices: respond to income, rates, migration, supply ---
    price = [price0]
    mig_mean, mig_std = net_migration.mean(), net_migration.std()
    perm_mean, perm_std = permits.mean(), permits.std()
    inv_mean, inv_std = inventory.mean(), inventory.std()

    for t in range(1, n):
        inc_g = income[t] / income[t - 1] - 1
        mort_chg = mort[t] - mort[t - 1]
        mig_z = (net_migration[t] - mig_mean) / mig_std
        perm_z = (permits[t] - perm_mean) / perm_std
        inv_z = (inventory[t] - inv_mean) / inv_std

        shock = np.random.normal(0, 0.01)

        g_price = (
            0.4 * inc_g      # income up -> prices up
            - 3.0 * mort_chg # rates up -> prices down
            + 0.01 * mig_z   # inflow -> demand up
            - 0.005 * perm_z # more building -> rel. cheaper
            - 0.003 * inv_z  # more inventory -> weaker price
            + shock
        )

        price.append(price[-1] * (1 + g_price))

    # collect rows
    for i, d in enumerate(dates):
        rows.append(
            dict(
                region=region,
                date=d,
                price=price[i],
                income=income[i],
                population=pop[i],
                net_migration=net_migration[i],
                rent=rent[i],
                mortgage_rate=mort[i],
                permits=permits[i],
                inventory=inventory[i],
                unemployment=unemployment[i],
            )
        )

# ---------------------------------------------------------------------
# 1. Build panel and split into individual CSVs
# ---------------------------------------------------------------------
panel = pd.DataFrame(rows)

panel["date"] = panel["date"].dt.strftime("%Y-%m-%d")

housing_price = panel[["region", "date", "price"]]
income = panel[["region", "date", "income"]]
population_migration = panel[["region", "date", "population", "net_migration"]]
mortgage_rate = panel[["region", "date", "mortgage_rate"]]
rent = panel[["region", "date", "rent"]]
permits = panel[["region", "date", "permits"]]
inventory = panel[["region", "date", "inventory"]]
unemployment = panel[["region", "date", "unemployment"]]

housing_price.to_csv(DATA_RAW / "housing_price.csv", index=False)
income.to_csv(DATA_RAW / "income.csv", index=False)
population_migration.to_csv(DATA_RAW / "population_migration.csv", index=False)
mortgage_rate.to_csv(DATA_RAW / "mortgage_rate.csv", index=False)
rent.to_csv(DATA_RAW / "rent.csv", index=False)
permits.to_csv(DATA_RAW / "permits.csv", index=False)
inventory.to_csv(DATA_RAW / "inventory.csv", index=False)
unemployment.to_csv(DATA_RAW / "unemployment.csv", index=False)

# ---------------------------------------------------------------------
# 2. Simple policy_events file (optional, not used by pipeline yet)
# ---------------------------------------------------------------------
events = []
for region in regions:
    for date, label in [
        ("2011-08-01", "easing"),
        ("2015-01-01", "macro_prudential_tightening"),
        ("2020-04-01", "covid_support"),
        ("2022-03-01", "rate_hike_cycle_start"),
    ]:
        events.append(
            dict(region=region, date=date, event_type=label)
        )

policy_events = pd.DataFrame(events)
policy_events.to_csv(DATA_RAW / "policy_events.csv", index=False)

print(f"Written synthetic CSVs to {DATA_RAW}")
```

```python
def compute_forward_return(df, horizon=12):
    """
    Compute forward price return over horizon months.
    """
    out = df.sort_values(["region", "date"]).copy()
    out["price_fwd"] = out.groupby("region")["price"].shift(-horizon)
    out["fwd_return"] = out["price_fwd"] / out["price"] - 1
    return out
```

```python
def assign_quintiles(df, signal_col):
    """
    Assign quintiles based on signal (per date).
    """
    def _qcut(x):
        return pd.qcut(x, 5, labels=False, duplicates="drop") + 1

    out = df.copy()
    out["quintile"] = out.groupby("date")[signal_col].transform(_qcut)
    return out
```

```python
def directional_backtest(df, signal_col="affordability_score", horizon=12):
    df = compute_forward_return(df, horizon=horizon)
    df = df.dropna(subset=["fwd_return"])

    df = assign_quintiles(df, signal_col)

    summary = (
        df.groupby("quintile")["fwd_return"]
        .mean()
        .to_frame("avg_fwd_return")
    )

    summary["rank"] = summary.index
    return summary
```

```python
df = run_pipeline("austin")

summary = directional_backtest(
    df,
    signal_col="composite_score",
    horizon=12
)

summary
```

```python
import numpy as np

def monthly_payment(price, rate, years=25):
    """
    Vectorized monthly mortgage payment.
    price: Series or float
    rate: Series or float (annual)
    """
    r = rate / 12
    n = years * 12

    payment = price * (r * (1 + r) ** n) / ((1 + r) ** n - 1)

    # handle zero-rate edge case safely
    payment = np.where(r == 0, price / n, payment)

    return payment

```
